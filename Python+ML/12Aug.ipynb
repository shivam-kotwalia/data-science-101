```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
X = np.linspace(-np.pi*3, np.pi*3, 256, endpoint=True)
```

```python
X.shape
```

```python
C = np.cos(X)
S = np.sin(X)
```

```python
plt.figure(figsize=(10,6), dpi=80)
plt.plot(X, C, label="Cosine")
plt.plot(X, S, label='Sin')
plt.legend(loc='upper left')
```
