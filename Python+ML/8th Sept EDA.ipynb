```python
first_q = np.percentile(df.pelvic_radius, 25)
third_q = np.percentile(df.pelvic_radius, 75)
```

```python
IQR = third_q - first_q
```

```python
IQR *= 1.5
```

```python
min_value = first_q - IQR
max_value = third_q + IQR
```

```python
min_value, max_value
```

```python
df[df.pelvic_radius > max_value]["pelvic_radius"]
```

```python
df[df.pelvic_radius < min_value]["pelvic_radius"]
```

```python
df[(df.pelvic_radius > max_value) | (df.pelvic_radius < min_value)]["pelvic_radius"]
```
