```python
big_list = []
label = []
for im_filename in os.listdir("../data/dog_cat_small_data/"):
    image = cv2.imread("../data/dog_cat_small_data/" + im_filename, 0)
    image = cv2.resize(image, (150, 150))
    big_list.append(image)
    if im_filename.startswith("dog"):
        label.append(1)
    else:
        label.append(0)
```

```python
len(big_list)
```

```python
import numpy as np 
```

```python
images = np.array(big_list)
```

```python
images.shape
```

```python
i = 91
plt.imshow(images[i])
print(images[i].shape)
```

```python
min_w = min_h = 100
max_h = max_w = 0 
for i in os.listdir("../data/dog_cat_small_data/"):
    image = cv2.imread("../data/dog_cat_small_data/" + i)
    shape = image.shape
    if shape[0] > max_h:
        max_h = shape[0]
    if shape[0] < min_h:
        min_h = shape[0]
    if shape[1] > max_w:
        max_w = shape[1]
    if shape[1] < min_w:
        min_w = shape[1]
```

```python
label = np.array(label)
```

```python
images.shape, label.shape
```

```python
label = keras.utils.to_categorical(label, num_classes=2)
```

```python
images.shape, label.shape
```

```python
test_label = [5,2,3,3,2,1,2,3,1,2,3,1,2,1,2,1,2,2,3,3,3,3]
```

```python
keras.utils.to_categorical(test_label)
```

```python
images = images/255.0
```

```python
import keras
```

```python
model = keras.models.Sequential()
model.add(keras.layers.Flatten())
#model.add(keras.layers.Dense(1024, activation="relu"))
#model.add(keras.layers.Dropout(0.4))
model.add(keras.layers.Dense(512, activation='relu'))
#model.add(keras.layers.Dropout(0.2))
model.add(keras.layers.Dense(2, activation='softmax'))
```

```python
model.compile(optimizer='SGD', loss=keras.losses.categorical_crossentropy, metrics=["accuracy"])
```

```python
model.fit(images, label, batch_size=128, epochs=10, verbose=1, validation_split=0.2)
```
