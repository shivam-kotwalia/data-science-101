```python
big_list = []
label = []
for im_filename in os.listdir("../data/dog_cat_small_data/"):
    image = cv2.imread("../data/dog_cat_small_data/" + im_filename)
    image = cv2.resize(image, (150, 150))
    big_list.append(image)
    if im_filename.startswith("dog"):
        label.append(1)
    else:
        label.append(0)
```

```python
len(big_list)
```

```python
import numpy as np 
```

```python
images = np.array(big_list)
```

```python
images.shape
```

```python
i = 91
plt.imshow(images[i])
print(images[i].shape)
```

```python
min_w = min_h = 100
max_h = max_w = 0 
for i in os.listdir("../data/dog_cat_small_data/"):
    image = cv2.imread("../data/dog_cat_small_data/" + i)
    shape = image.shape
    if shape[0] > max_h:
        max_h = shape[0]
    if shape[0] < min_h:
        min_h = shape[0]
    if shape[1] > max_w:
        max_w = shape[1]
    if shape[1] < min_w:
        min_w = shape[1]
```
